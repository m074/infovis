```
import pandas as pd
from datetime import datetime, timedelta
import csv

START_DATE = datetime.strptime('2019-09-21 22:25:48', '%Y-%m-%d %H:%M:%S')

mem = START_DATE
flag_on = 1

df = pd.read_csv('https://raw.githubusercontent.com/m074/infovis/master/pdata/data.csv')

acum_day = 0
acum_hour = 0
flag_on = 1
day_dict = {'date': [], 'value': []}
hour_dict = {'date': [], 'hour': [], 'value': []}

for row in df.values:
    id, value, feed_id, date, *_ = row
    datetime_object = datetime.strptime(date, '%Y-%m-%d %H:%M:%S')
    datetime_local = datetime_object


    while mem <= datetime_local:
        last_day = mem.weekday()
        last_hour = mem.hour
        mem += timedelta(seconds=60)

        if last_hour != mem.hour:
            hour_dict['date'] += [str((mem - timedelta(seconds=60)).date())]
            hour_dict['hour'] += [(mem - timedelta(seconds=60)).hour]
            hour_dict['value'] += [acum_hour/60]
            acum_hour = 0

        if last_day != mem.weekday():
            day_dict['date'] += [str((mem-timedelta(seconds=60)).date())]
            day_dict['value'] += [acum_day/60]
            acum_day = 0

        acum_day += flag_on
        acum_hour += flag_on

    if value == 'OFF':
        flag_on = 0
    else:
        flag_on = 1


day_data = pd.DataFrame(day_dict)
hour_data = pd.DataFrame(hour_dict)

print('cargado')
```
